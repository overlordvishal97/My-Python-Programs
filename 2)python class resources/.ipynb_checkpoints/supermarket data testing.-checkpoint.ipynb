```python
from geopy.geocoders import ArcGIS
```

```python
nom =ArcGIS()
```

```python
#this geocode gives the coordinates of the address we give.sometimes it does give nothing then we have to assume the address does not exist something like that.
#if  we put in a variable with a false address then it clearly displays 'None'.
nom.geocode("3995 23rd St,San Francisco,CA 94114")
```

```python
#it even shows the coordinates of india.
nom.geocode("Srinivasa nagar,kurnool,India,518004")
```

```python
df=pandas.read_csv("supermarkets.csv")
df
```

```python
df["Address"]+", "+["San Francisco "]+", "+["USA "]
df
```

```python
df["Coordinates"]=df["Address"].apply(nom.geocode)
df.Coordinates
```

```python
#you particularly slice coordinates from the table and even select latitudes are longitudes from the coordinates.
df.Coordinates[0].latitude
```

```python
df=df.drop("Coordinate",1)
df
```

```python
#lambda is an inline function to build your own function.
df["Latitude"]=df["Coordinates"].apply(lambda x: x.latitude)
df
```

```python
#if it ever shows none the use if x != none else none
df["Longitude"]=df["Coordinates"].apply(lambda x: x.longitude)
df
```
